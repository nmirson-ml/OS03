```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy import stats
import latexify 
import math
```

```python
p = .9  #  probability of each trial
m = 1000 # number of trials
#switchs = [a,b,c,d,e]

# bernoulli is binomial with n = 1
a = np.random.binomial(1, p, m)
b = np.random.binomial(1, p, m)
c = np.random.binomial(1, p, m)
d = np.random.binomial(1, p, m)
e = np.random.binomial(1, p, m)

# four different paths
path_1 = a * d
path_2 = a * c *e
path_3 = b * e
path_4 = b * c * d
```
